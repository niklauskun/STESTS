```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Aggregated BESS

```python
# Creating the DataFrame from the provided data
data = {
    "EDH": [1, 1, 1, 6, 6, 6, 13, 13, 13],
    "Segments": [1, 5, 10, 1, 5, 10, 1, 5, 10],
    "Computation Time (ms)": [4.948327584, 5.817620527, 6.329186024, 36.79917055, 42.00672287, 46.08614246, 92.4571412, 101.7240094, 107.1902953]
}

df = pd.DataFrame(data)
df['EDH'] = df['EDH'].astype(str) + '-RTM Intervals'
```

```python
# Plotting
sns.set(style="whitegrid", font_scale=1.4)
plt.figure(figsize=(14, 8))

# Set the palette using the Set2 colormap
palette = sns.color_palette("coolwarm")

# Create a bar plot with the Set2 colormap
sns.barplot(x='EDH', y='Computation Time (ms)', hue='Segments', data=df, palette=palette)

# Title and labels
plt.title('Computation Time for Economic Dispatch Problems (Aggregated BESS)', fontsize=20)
plt.xlabel('Time Horizon', fontsize=18)
plt.ylabel('Computation Time (ms)', fontsize=18)
plt.ylim([0, 140])


# Customize the legend
plt.legend(title='Market Model Segments', title_fontsize='16', fontsize='14', loc='upper left')

# Show plot
plt.show()
```

```python
# Plotting with the new x-axis as Total BESS Unit and y-axis as Computation Time (ms) using ED (ms/period)
plt.figure(figsize=(14, 8))

# Colors for different EDH values
colors = {'1': 'blue', '13': 'red', '13-SoC': 'green'}
# Markers for different Seg values
markers = {1: 'o', 5: '^', 10: 's'}

# Plot for each Seg value, showing the effect of Total BESS Unit for different EDH values
for edh in sorted(df['EDH'].unique()):
    for seg in df['Seg'].unique():
        subset = df[(df['Seg'] == seg) & (df['EDH'] == edh)]
        plt.plot(subset['Total BESS Unit'], subset['ED (ms/period)'], 
                 marker=markers[seg], color=colors[edh], linestyle='-', 
                 label=f'Seg = {seg}, EDH = {edh}', markersize=10)

plt.xlabel('Total BESS Unit', fontsize=20)
plt.ylabel('Computation Time (ms)', fontsize=20)
plt.title('Impact of Seg, EDH, and Total BESS Unit on Computation Time (ms)', fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

# Sort legend by EDH and then by Seg
handles, labels = plt.gca().get_legend_handles_labels()
sorted_labels_handles = sorted(zip(labels, handles), key=lambda x: (x[0].split('=')[2].split(',')[0], int(x[0].split('=')[1].split(',')[0])))
sorted_labels, sorted_handles = zip(*sorted_labels_handles)
plt.legend(sorted_handles, sorted_labels, fontsize=16, ncols=3)

plt.grid(True)
plt.tight_layout()
plt.show()
```
